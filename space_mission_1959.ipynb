```python
import pandas as pd
```

```python
df=pd.read_csv("Space_Corrected.csv")
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Detail'].value_counts()[:20]
```

```python
df["Country"]=df['Location'].str.split(",").str[-1]

df
```

```python
df['Country'].value_counts()
```

```python
df=df.drop(["Unnamed: 0", "Unnamed: 0.1"],axis=1)
df
```

```python

df['Launch_time']=df['Datum'].str.split(",").str[-1]
df['Launch_date']=df['Datum'].str.split(", ").str[-2]
```

```python
df
```

```python
df['day']=df['Launch_date'].str.split().str[0]
df['month']=df['Launch_date'].str.split().str[1]
df['year']=df['Launch_time'].str.split().str[0]
df.head()
```

```python
df['time']=df['Launch_time'].str.split().str[1]
df
```

```python
df['Launch_hour']=pd.to_datetime(df.time).dt.hour
df['Launch_minute']=pd.to_datetime(df.time).dt.minute
df=df.drop(['Datum', 'Launch_time', 'Launch_date', 'time'], axis=1)
df
```

```python
df['Country'].value_counts()
```

```python
df['Launch_hour'].fillna(df['Launch_hour'].mean(), inplace=True)
df['Launch_minute'].fillna(df['Launch_minute'].mean(), inplace=True)
```

```python
df['Status Mission'].value_counts()
```

```python
df['Detail'].value_counts()[:10]
```

```python
df['satellite_name']=df['Detail'].str.split('|').str[1]
df['Satellite_name']=df['satellite_name'].str.split('&').str[0]
df['Launch_system']=df['Detail'].str.split('|').str[0]
```

```python
df.isnull().sum()
```

```python
df=df.drop([' Rocket', 'Location','Detail','satellite_name'], axis=1)
df.head(5)
```

```python
df["Satellite_name"].value_counts()
```

```python
Company_name=df['Company Name']
Company_name=pd.get_dummies(df['Company Name'], drop_first=True)
```

```python
Company_name
```

```python
df['Status Rocket'].value_counts()
```

```python
df['Status Rocket'].replace({'StatusRetired':0, 'StatusActive':1}, inplace=True)
df['day'].replace({'Sun':0, 'Mon':1, 'Tue':2, 'Wed':3, 'Thu':4, 'Fri':5, 'sat':6, 'Sat':6}, inplace=True)
```

```python
df['month'].value_counts()
```

```python
df['month'].replace({'Jan':0, 'Feb':1, 'Mar':2, 'Apr':3, 'May':4, 'Jun':5, 'Jul':6, 'Aug':7, 'Sep':8, 'Oct':9, 'Nov':10, 'Dec':11}, inplace=True)
```

```python
country=pd.get_dummies(df['Country'], drop_first=True)
country.info()
```

```python
country
```

```python
Satellite_name=df['Satellite_name']
Satellite_name=pd.get_dummies(Satellite_name, drop_first=True)

```

```python
Launch_system=df['Launch_system']
Launch_system=pd.get_dummies(Launch_system, drop_first=True)
```

```python
df
```

```python
df=pd.concat([df,Company_name, country, Satellite_name, Launch_system], axis=1)
```

```python
df
```

```python
df=df.drop(['Company Name','Country', 'Satellite_name','Launch_system'], axis=1)
```

```python

df['Status Mission'].value_counts()
```

```python
df['Status Mission'].replace({'Success':1, 'Failure':0, 'Partial Failure':0, 'Prelaunch Failure':0}, inplace=True)
```

```python
df
```

# Deal with Imabalanced dataset

2. RandomOverSampler

# 1.SMOTE 

```python
from imblearn.combine import SMOTETomek
from imblearn.under_sampling import NearMiss
```

```python
smk=SMOTETomek(random_state=42)
X_smote, y_smote=smk.fit_sample(X,y)
```

```python
print(X_smote.shape)
print(y_smote.shape)
```

```python
from collections import Counter
```

```python
print("Original Dataset is {}".format(Counter(y)))
print("Resample Dataset is {}".format(Counter(y_smote)))
```

```python
X_smote_train, X_smote_test, y_smote_train, y_smote_test=train_test_split(X_smote, y_smote, train_size=0.2, random_state=0)
print(X_smote_train.shape)
print(X_smote_test.shape)
print(y_smote_train.shape)
print(y_smote_test.shape)
```

# RandomForestClassifier

```python
classifier=RandomForestClassifier()
classifier.fit(X_smote_train, y_smote_train)
classifier_score=classifier.score(X_smote_train, y_smote_train)
print(classifier_score)
```

```python
classifier_score_test=classifier.score(X_smote_test, y_smote_test)
print(classifier_score_test)
```

```python
prediction=classifier.predict(X_smote_test)
```
